### Task 7




##### List the total number of successful and failure mission outcomes

### Task 8



##### List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
# Find the maximum payload mass
max_payload = df["PAYLOAD_MASS__KG_"].max()
print(f"Maximum payload mass: {max_payload} kg")

# Find all booster versions that carried the maximum payload mass
boosters_with_max_payload = df[df["PAYLOAD_MASS__KG_"] == max_payload]["Booster_Version"].unique()
print(f"Booster versions with maximum payload mass:")
for booster in boosters_with_max_payload:
    print(f"- {booster}")
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
# Convert Date column to datetime if it's not already
df['Date'] = pd.to_datetime(df['Date'])

# Filter for 2015 and failure landing outcomes in drone ship
filtered_df = df[(df['Date'].dt.year == 2015) & 
                 (df['Landing_Outcome'] == 'Failure (drone ship)')]

# Create a copy to add month names
result_df = filtered_df.copy()
result_df['Month'] = filtered_df['Date'].dt.month_name()

# Display the required columns
columns_to_display = ['Month', 'Landing_Outcome', 'Booster_Version', 'Launch_Site', 'Date']
print("Records for failure landing outcomes in drone ship for 2015:")
print(result_df[columns_to_display])
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
# Convert Date column to datetime if not already done
df['Date'] = pd.to_datetime(df['Date'])

# Filter data between the specified date range
start_date = '2010-06-04'
end_date = '2017-03-20'
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

# Count landing outcomes and sort in descending order
landing_outcome_counts = filtered_df['Landing_Outcome'].value_counts().sort_values(ascending=False)

print(f"Landing outcome counts between {start_date} and {end_date} (ranked in descending order):")
for i, (outcome, count) in enumerate(landing_outcome_counts.items(), 1):
    print(f"{i}. {outcome}: {count}")
```
